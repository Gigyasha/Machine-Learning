# Applying Naive Baye's Classifier on a Categorical Dataset

Naive bayes Equation<br>
\begin{equation}
 p(A|B) = \frac{P(B \cap A)}{p(B)}
\end{equation}
<br>
when we deal with applying naive bayes classifier on data which has categorical features then we cant use the normal pdf<br>
if this dataset the stalk root column has ? in the column so to fill such values with other values I have creates a sample<br>distribution from the stalk-root column (without ? values) and did a sampling from that so to fill up ? values<br>

```python
prior_p_train = p_training_data.shape[0]/Training_data.shape[0]
prior_e_train = e_training_data.shape[0]/Training_data.shape[0]
print(prior_p_train,prior_e_train)
```

when checking the values in e_training_data and in p_training_data we can see for the columns the unique value differs like the cap-shape column in training data has 6 unique values but in e_training_data and in p_training_data the unique value<br>for same column is diiefrent in no like cap_shape_e =[x,b,f,k] and cap_shape_p=[b,s,k,c,f] and the all values in <br>training data are =[x,b,s,f,k,c] so we can see that c in missing in cap_shape_e but we cant put the probability for <br>that as 0 beacuse it may not be present in e_training_data but might be present in cross validation data for class e for for such <br>condition we apply __LAPLACE SMOOTHENING__

```python
def get_probabilites(new_data):
    final_dict=dict()
    for col in data.columns[1:]:
        if len(data[col].unique())!=len(new_data[col].unique()):
        
            not_present = set(data[col].unique())-set(new_data[col].unique())
            #print(list(not_present))
            sub_dict=dict()
            for no in list(not_present):
                sub_dict[no]=1/(len(new_data)+len(data[col].unique()))
            #print(sub_dict)
            dict1 = dict((new_data[col].value_counts())/new_data.shape[0])
            dict1.update(sub_dict)
            
            final_dict[col] = dict1
         
        else:
             final_dict[col] = dict((new_data[col].value_counts())/new_data.shape[0])
    return final_dict    
```

```python
p_all_prob = get_probabilites(p_training_data)
```

```python
e_all_prob = get_probabilites(e_training_data)
```

```python
def posterior_probabilities(x):
    final_list=[]
    for ind in x.index:
        if i!=0:
            df = list(x.loc[ind])
            liklihood_p=1.0
            liklihood_e=1.0
            for d,col in zip(df,Cross_validation_data.columns[1:]):
                liklihood_p = liklihood_p  * p_all_prob[col][d]
                liklihood_e = liklihood_e  * e_all_prob[col][d]
            #print(liklihood_p,liklihood_e)
            posterior_p =(liklihood_p*prior_p_train)/((liklihood_p*prior_p_train)+(liklihood_e*prior_e_train))
            final_list.append(posterior_p)
       
    return np.array(final_list)
    #print(x[1])
    #liklihood_p =
    #liklihood_e
    
```
