replacing the na values with nan and after that fiiling the nan value by the most frequest value of that column

```python
def remove_na_values(col):
    
    data[col].replace(to_replace="na",value=np.NaN,inplace=True)  #replacing na to nan so that we can apply fill na
    
    values_dict = dict((data[col].value_counts()))
    
    #print(values_dict)
    
    key = list(values_dict.keys())
    
    most_freq_val = key[0]
    
    data.fillna(value=most_freq_val,inplace=True)
```

```python
def reduced_dimenion(factorized_matrix,variance_to_preserve):
    
    eig_val_sum=sum(factorized_matrix[1])
    
    e_val=0
    
    L=0
    
    for eig_val in factorized_matrix[1]:
        
        e_val=e_val+eig_val
        
        if e_val/eig_val_sum>=variance_to_preserve:
            break
        
        
        L+=1
    return L
```

## Code For SMOTE

In below cell k=20 it means 1 vector is used to generate 20 fake data so 1000 will help us in generating 20000 synthetic data if one want to reduce or increase the no it cab be done by changing K.

```python
pos_class_features = pos_class_data.iloc[:,:-1]
pos_class_features = np.array(pos_class_features)
pos_class_features.shape
```

```python
g=0
```

```python
def smote(r_vector,temp_pos_data,k):
    
    for i in range(k):
        global g
        
        a_index = np.argmin(np.linalg.norm((r_vector - temp_pos_data),ord=2,axis=1))
        
        a_vector = temp_pos_data[a_index]
        
        a_vector = a_vector.reshape(1,a_vector.shape[0])
        
        g+=1
        
        #print(a_vector.shape)
        
        temp_pos_data = np.delete(temp_pos_data,obj=a_index,axis=0)
        
        print("Generated--> ",g)
        
        #print(temp_pos_data.shape)
        
        if i ==0:
            
            a_vectors=a_vector
            
        else:
            
            a_vectors = np.concatenate((a_vectors,a_vector),axis=0)
        
    return a_vectors  # this will return the k most closest vector to r
```

```python
k=20

for j in range(pos_class_data.shape[0]):
    
    index = np.random.randint(0,high=(pos_class_features.shape[0])-1)

    r_vector = pos_class_features[index]

    r_vector = r_vector.reshape(1,r_vector.shape[0])

    index_to_remove = np.argmin(np.linalg.norm((r_vector - pos_class_features),ord=2,axis=1))

    temp_pos_data = np.delete(pos_class_features,obj=index_to_remove,axis=0)

    r_vector.shape
    
    temp_pos_data.shape
    
    a_vectors = smote(r_vector,temp_pos_data,k)
    
    #print(a_vectors.shape)

    b_vectors = r_vector - a_vectors   #finding b so that we can teduce its size

    s_vectors = r_vector + np.random.uniform(0,1,size=1)*b_vectors   #reducing the size of b and adding t that (r+b')=s vector
    
    if j==0:
        
        synthetic_pos_features = s_vectors
        
    else:
        
        synthetic_pos_features = np.concatenate((synthetic_pos_features,s_vectors),axis=0)
    
```

Now calculating the parameters s that we can perform classification

```python
pos_train_data.iloc[:,:-1]
```

```python
pos_train_data = pos_train_data.iloc[:,:-1].astype("float")
neg_train_data = neg_train_data.iloc[:,:-1].astype("float")
```

```python
cv_data = cv_data.iloc[:,:-1].astype("float")
```

```python
pos_train_data.shape,neg_train_data.shape
```

```python
pos_train_data.head()
```

```python
prob_pos = pos_train_data.shape[0]/training_data.shape[0]
prob_neg = neg_train_data.shape[0]/training_data.shape[0]
print("probabilites for positive class and probabilites of neg classes",prob_pos,prob_neg)
```

```python
mu_cap_pos = np.array((pos_train_data).mean())
mu_cap_neg = np.array((neg_train_data).mean())
cov_pos = np.array((pos_train_data).cov())
cov_neg = np.array((neg_train_data).cov())
print("Parameters for positive class-->",mu_cap_pos,cov_pos)
print("Parameters for negative class-->",mu_cap_neg,cov_neg)
```

```python
#mu_cap_pos = mu_cap_pos.reshape()
mu_cap_neg.shape,mu_cap_pos.shape
```

```python
import scipy.stats as s
```

```python
def classification(X):
    
    P_liklihood = s.multivariate_normal.pdf(X,mean=mu_cap_pos,cov=cov_pos)*prob_pos
    
    N_liklihood = s.multivariate_normal.pdf(X,mean=mu_cap_neg,cov=cov_neg)*prob_neg
    
    Numerator = P_liklihood
    
    Denominator = P_liklihood+N_liklihood
    
    return (Numerator/Denominator)  #return the prob for 1 or pos class
```
